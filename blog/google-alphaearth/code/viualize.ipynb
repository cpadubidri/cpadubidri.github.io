```python
import numpy as np

N, D = 5, 128
X = np.random.randn(N, D)

# 1) Gram matrix
G = X @ X.T  # (N, N)

# 2) Cosine similarity
norms = np.linalg.norm(X, axis=1, keepdims=True)
S_cos = (X @ X.T) / (norms @ norms.T + 1e-12)

print("Gram matrix:\n", G)
print("Cosine similarity:\n", S_cos)
```

```python
import torch
import torch.nn.functional as F

# Suppose you already have a batch of embeddings from a model
# shape: (N, D) = (num_items, embedding_dim)
# e.g., random demo data:
N, D = 5, 128
X = F.normalize(torch.randn(N, D), p=2, dim=1)  # often embeddings are L2-normalized

# 1) Gram matrix (inner products): G_ij = <x_i, x_j>
G = X @ X.t()            # shape: (N, N)

# 2) Cosine-similarity matrix: with L2-normalized rows, Gram == cosine
# If X wasn't normalized, you can do:
X_raw = torch.randn(N, D)
X_unit = F.normalize(X_raw, p=2, dim=1)
S_cos = X_unit @ X_unit.t()

# 3) If you want Euclidean distances too (not a Gram matrix, but common for similarity):
D_euc = torch.cdist(X, X, p=2)  # shape: (N, N)

print("Gram matrix:\n", G)
print("Cosine similarity:\n", S_cos)
print("Euclidean distances:\n", D_euc)
```
